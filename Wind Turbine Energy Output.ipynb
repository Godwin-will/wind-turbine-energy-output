## Determine the Energy Output of a Wind Turbine

### Defining Parameters

```python
#Air Density = µ (kg/m^3)
#Diameter = D (m)                   
#Speed = S (m/s)                        
#Efficiency = å
#Wind Power = WP
#Energy Output = EO
```

### Energy Output vs Wind Power

```python
import numpy as np 
import matplotlib.pyplot as mango
%matplotlib inline

D = np.random.randint(8,57,7)
S = np.random.randint(6,80,7)
å = 0.40
µ = 1.225
π = np.pi

for Diameter in D: 
    A = (1/4*π*D**2)
    for Area in A: 
        AV = (1/2*µ*A)
        for Speed in S:
            WP = (AV*S**3) 
            for Wind_Power in WP:
                EO = (WP*å)

mango.figure(figsize=(9, 5))
mango.plot(WP, EO, 'r^')
mango.grid(True)
mango.xlabel('Wind Power', fontsize=12, color='b')
mango.ylabel('Wind Turbine Energy Output', fontsize=12, color='b')
mango.title('Energy Output vs Wind Power', fontsize=18, color='b')
```
